```python
data = pd.read_csv("../input/weatherww2/Summary of Weather.csv")
```

```python
data.head()
```

```python
sns.pairplot(data , x_vars = ['MinTemp'] , y_vars = ['MaxTemp'] , size = 7 , aspect = 0.7 , kind = 'reg')
```

```python
x_data = data['MinTemp']
y_data = data['MaxTemp']

x = np.array(x_data).reshape(-1, 1)
y = np.array(y_data)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,random_state = 1)
```

```python
print(X_test.shape)
```

```python
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
print(linreg.intercept_)
```

```python
print(linreg.coef_)
```

```python
#drawing training
plt.scatter(X_train, y_train, color= 'black')

plt.plot(X_train, linreg.predict(X_train), color = 'yellow')

plt.title ("Training Dataset")

plt.xlabel("Minimum Temperature")

plt.ylabel("Maximum Temperature")

plt.show()


#drawing test
plt.scatter(X_test, y_test, color= 'black')

plt.plot(X_train, linreg.predict(X_train), color = 'yellow')

plt.title("Testing Dataset")

plt.xlabel("Minimum Temperature")

plt.ylabel("Maximum Temperature")

plt.show()
```

```python
import numpy
#finding the mean
x_mean = numpy.mean(x)
print('x_mean :',x_mean)
print()

y_mean = numpy.mean(y)
print('y_mean :',y_mean)
print()


#finding the median
x_median = numpy.median(x)
print('x_median :',x_median)
print()

y_median = numpy.median(y)
print('y_median :',y_median)
print()

from scipy import stats

#finding the mode
x_mode = stats.mode(x)
print('x_mode :',x_mode)
print()

y_mode = stats.mode(y)
print('y_mode :',y_mode)
print()

#finding the standard deviation
x_stdev = numpy.std(x)
print('standard deviation(x)',x_stdev)
print()

y_stdev = numpy.std(y)
print('standard deviation(y)',y_stdev)
print()

#finding the variance
x_var = numpy.var(x)
print('variance(x)',x_var)
print()
y_var = numpy.var(y)
print('variance(y)',x_var)
print()
```
